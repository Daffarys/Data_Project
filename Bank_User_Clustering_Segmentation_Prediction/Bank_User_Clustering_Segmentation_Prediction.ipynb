# Check Multicolinerity

```python
def calc_vif(X):
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

Based on these data there are about 10 column with a multicolinierity score of below 5, since the column cash advance frequency have a score of 4.82 which is close to 5 therefore we decide to exclude it for this argument which makes it that we can use about 8 columns or features with cash advance frequency included

```python
# Dimensionality Reduction using PCA


pca = PCA()
pca.fit(X_final)

cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1

plt.figure(figsize=(6,4))
plt.plot(cumsum, linewidth=3)
plt.axis([0, 10, 0, 1])
plt.xlabel("Dimensions")
plt.ylabel("Cumulative of Explained Var Ratio")
plt.plot([d, d], [0, 0.95], "k:")           
plt.plot([0, d], [0.95, 0.95], "k:")        
plt.plot(d, 0.95, "ko")                    
plt.grid(True)
plt.show()
```

# Elbow Method

```python
inertias = []
sil_score = []
for k in range(2,21):
  km = KMeans(n_clusters=k).fit(X_final_pca)
  inertias.append(km.inertia_)
  sil_score.append(silhouette_score(X_final_pca,km.labels_))
```

# Silhouette Plot

```python
# Function to create Silhouette Plot

def plot_silhouette(range_n_clusters, X, random_state):
    for n_clusters in range_n_clusters:
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig.set_size_inches(18, 4)

      
        ax1.set_xlim([-0.1, 1])

        ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

        clusterer = KMeans(n_clusters = n_clusters, init = 'k-means++', max_iter = 300, n_init = 10, random_state = random_state)
        cluster_labels = clusterer.fit_predict(X)


        silhouette_avg = silhouette_score(X, cluster_labels)
        print("For n_clusters =", n_clusters,
              "The average silhouette_score is :", silhouette_avg)

        sample_silhouette_values = silhouette_samples(X, cluster_labels)

        y_lower = 10
        for i in range(n_clusters):

            ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
            ith_cluster_silhouette_values.sort()

            size_cluster_i = ith_cluster_silhouette_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_clusters)
            ax1.fill_betweenx(np.arange(y_lower, y_upper),
                              0, ith_cluster_silhouette_values,
                              facecolor=color, edgecolor=color, alpha=0.7)


            ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))


            y_lower = y_upper + 10  

        ax1.set_title("The silhouette plot for the various clusters.")
        ax1.set_xlabel("The silhouette coefficient values")
        ax1.set_ylabel("Cluster label")


        ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

        ax1.set_yticks([]) 
        ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

      
        colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
        ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                    c=colors, edgecolor='k')

        centers = clusterer.cluster_centers_

        ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                    c="white", alpha=1, s=200, edgecolor='k')

        for i, c in enumerate(centers):
            ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1, s=50, edgecolor='k')

        ax2.set_title("The visualization of the clustered data.")
        ax2.set_xlabel("Feature space for the 1st feature")
        ax2.set_ylabel("Feature space for the 2nd feature")

        plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                      "with n_clusters = %d" % n_clusters),
                     fontsize=14, fontweight='bold')
```

```python
# Display the Silhouette Score and Plot Silhouette Plot

plot_silhouette(range(2, 10), X_final_pca, random_state=123)
```

```python
# Display the Silhouette Score and Plot Silhouette Plot

fig, axes = plt.subplots(5, 3, figsize=(15, 10))

# Loop for each plot
for i, ax in enumerate(axes.flatten()):

    # Initialize K-means clustering
    n_clusters = i + 2
    kmeans = KMeans(n_clusters = n_clusters, init = 'k-means++', max_iter = 300, n_init = 10)

    # Create SilhouetteVisualizer
    visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick', ax=ax)

    # Fit visualizer to data
    visualizer.fit(X_final_pca)

    # Set title
    ax.set_title(f'K = {n_clusters}')

# Adjust layout
plt.tight_layout()
plt.show()
```

From the result, the best number of of cluster provided by the silhouette plot sits between 5 and 6. Which fits with the previously used elbow method, for this process we decide to use 5 cluster since the distribution from the visualization is more even than with 6 distribution

# Conceptual Problem

1. Inertia in the K-Means algorithm refers to a metric used to measure how far the data points in a cluster are from the corresponding centroid.Inertia is one of the important metrics used in the K-Means algorithm to measure how well clusters have been formed and to assist in the selection of the optimal number of clusters. The smaller the inertia value, the better the quality of clusters produced by the K-Means algorithm.

2.
- The Elbow Method is one of the techniques used in selecting the optimal number of clusters in the K-Means algorithm or other clustering methods. Its purpose is to help determine the most appropriate or optimal number of clusters for the given data. 
- To employ the elbow method for selection, you begin by creating a graph of Within-Cluster Sum of Squares (WCSS), which represents the cost function across various K values. WCSS quantifies the overall variance within each cluster, diminishing as K rises due to smaller clusters accommodating data points more closely. The key is to identify a distinct "elbow point" on the chart where the decline in WCSS reduction rate begins to decelerate, suggesting a potential optimal K.
- As for the advantages of the k means method, namely: Simple and easy to understand, Useful for initial exploration in understanding data structures.
- As for the disadvantages, there are: Sometimes the elbow curve is not so clear, and it is difficult to determine with certainty the optimal number of clusters.
The Elbow method only provides a rough guide, and in some cases, the optimal number of clusters may require further evaluation or other more sophisticated selection methods.
